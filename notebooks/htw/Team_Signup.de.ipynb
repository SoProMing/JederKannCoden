# Anleitung zur Prüfungsgruppenanmeldung

<img src="Team_Signup.jpeg" style="width:280px; height:280px;">

Diese Anleitung erklärt, wie du mit deiner Gruppe eine `.md`-Datei im GitLab-Repository erstellst, um euch für die Prüfung anzumelden. Du arbeitest dabei mit **Visual Studio 2022** und **GitLab**.

## 🔐 1. Zugriff auf das Repository

Beantrage Zugriff, wenn du ihn noch nicht hast:

![Zugriff anfragen](Team_Signup/Screenshot1.png)

## 🔗 2. Repository-URL kopieren

Gehe ins Repository und kopiere die HTTPS-URL:

![URL kopieren](Team_Signup/Screenshot3.png)

## 🧰 3. Repository in Visual Studio klonen

Öffne Visual Studio 2022 und klicke auf **Repository klonen**:

![Startansicht](Team_Signup/Screenshot2.png)

Füge die Repository-URL ein:

![Klondialog](Team_Signup/Screenshot4.png)

## 📂 4. Bestehende Datei (README.md) öffnen

Prüfe die Vorlage im Projekt:

![README öffnen](Team_Signup/Screenshot5.png)

## 🌱 5. Neuen Branch erstellen

Klicke unten rechts auf den Branchnamen main, wähle **Neuer Branch**, gib z. B. den Teamnamen ein:

![Branch wechseln](Team_Signup/Screenshot6.png)
![Branch erstellen](Team_Signup/Screenshot7.png)

## 📝 6. Neue Datei im Branch erstellen

Erstelle eine neue Markdown-Datei mit eurem Teamnamen:

![Neue Datei anlegen](Team_Signup/Screenshot8.png)

## ✍️ 7. Datei befüllen

Fülle die Datei nach Vorlage:

```markdown
# SoSe25_Prüfung

## Mitglieder
Alice Adams, Bob Boyle, Chris Carter, Dana Deutsch, Eva Evergreen

## Teamname
Test_123

## PZR
PZR 1
```

![Bearbeiten](Team_Signup/Screenshot9.png)

## 💾 8. Commit vorbereiten und durchführen

Trage eine Commit-Nachricht ein:

![Commit vorbereiten](Team_Signup/Screenshot10.png)

Wähle „**gestagte committen und synchronisieren**“:

![Commit & Push](Team_Signup/Screenshot11.png)

![Merge Request Hinweis](Team_Signup/Screenshot12.png)

## 🚀 9. Merge Request erstellen

Klicke in **gitlab** auf **Create merge request** und überprüfe die Änderungen:

![Merge Request Formular](Team_Signup/Screenshot13.png)

## 🔁 Optional: Manuelles Push aus Visual Studio

Falls du separat pushen möchtest:

![Push manuell](Team_Signup/Screenshot14.png)

## ✅ Ergebnis

Die Datei erscheint im Repository und ist korrekt eingecheckt:

![Finale Ansicht](Team_Signup/Screenshot15.png)

## 📌 Hinweise

- Der Branchname soll eindeutig und nachvollziehbar sein.
- Trage alle Gruppenmitglieder ein.
- Erstellt **einen Merge Request pro Gruppe**.
