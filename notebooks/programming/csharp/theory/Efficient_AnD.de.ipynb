QuickSort ist ein sehr effizienter, rekursiver Sortieralgorithmus, der das "Teile und herrsche"-Prinzip nutzt. #### Beispiel: QuickSort (rekursiv in C#)

```csharp
public static void QuickSort(int[] array, int left, int right)
{
    if (left >= right) return;

    int pivot = array[(left + right) / 2];
    int index = Partition(array, left, right, pivot);
    QuickSort(array, left, index - 1);
    QuickSort(array, index, right);
}

private static int Partition(int[] array, int left, int right, int pivot)
{
    while (left <= right)
    {
        while (array[left] < pivot) left++;
        while (array[right] > pivot) right--;
        if (left <= right)
        {
            (array[left], array[right]) = (array[right], array[left]);
            left++;
            right--;
        }
    }
    return left;
}
```

### Radix Sort

Hier ist eine vollst√§ndige, kommentierte **Radix Sort**-Implementierung in **C#**, die auf Ganzzahlen basiert. Radix Sort funktioniert am besten bei positiven Ganzzahlen und nutzt die **Ziffernpositionen** (Einer, Zehner, Hunderter usw.), um mehrfach zu sortieren ‚Äì jeweils mit einem stabilen Verfahren wie Counting Sort.

#### Beispiel: Radix Sort in C\#

```csharp

{
    static void Main()
    {
        int[] array = { 170, 45, 75, 90, 802, 24, 2, 66 };

    // Ziffernbasierter Counting Sort (stabil)
    private static void CountingSortByDigit(int[] array, int exp)
    {
        int n = array.Length;
        int[] output = new int[n];
        int[] count = new int[10]; // Ziffern 0-9

        // Z√§hle H√§ufigkeit jeder Ziffer an aktueller Stelle
        for (int i = 0; i < n; i++)
        {

        }

        // Errechne kumulierte Positionen
        for (int i = 1; i < 10; i++)

            count[i] += count[i - 1];
        }

        // Baue das Output-Array (von hinten f√ºr Stabilit√§t)
        for (int i = n - 1; i >= 0; i--)
        {
            int digit = (array[i] / exp) % 10;
            output[count[digit] - 1] = array[i];
            count[digit]--;
        }

        // Kopiere zur√ºck ins Originalarray
        for (int i = 0; i < n; i++)
        {
            array[i] = output[i];
        }
    }

    // Hilfsfunktion: finde das gr√∂√üte Element
    private static int GetMax(int[] array)
    {
        int max = array[0];
        foreach (int num in array)
        {
            if (num > max) max = num;
        }
        return max;
    }

    // Ausgabehilfe
    private static void PrintArray(int[] array)
    {
        foreach (int num in array)
        {
            Console.Write(num + " ");
        }
        Console.WriteLine();
    }
}
```

* **Laufzeit**:

  * O(d¬∑(n + k)), wobei `d` die Anzahl der Stellen, `n` die Anzahl der Elemente, und `k` der Zahlenbereich der Ziffern ist (hier 0‚Äì9 ‚Üí konstant).
  * F√ºr feste `d` ‚âà O(n).

```csharp
{
    static void Main()
    {

                {
                    // Tausche, wenn falsch sortiert
                    (array[j], array[j + 1]) = (array[j + 1], array[j]);
                    swapped = true;
                }
            }

            // Wenn kein Tausch mehr erfolgt ‚Üí sortiert
            if (!swapped)

        }
    }

    // üñ® Hilfsfunktion zum Anzeigen des Arrays
    public static void PrintArray(int[] array)

        {
```

### Programm: Vergleich von Sortieralgorithmen mit 1 Mio. Testdaten

````polyglot-notebook
using System;
using System.Diagnostics;

const int N = 1_000_000;

// Testdaten generieren
int[] dataForQuickSort = GenerateRandomArray(N, seed: 42);
int[] dataForRadixSort = (int[])dataForQuickSort.Clone();
int[] dataForBubbleSort = GenerateRandomArray(100_000, seed: 42); // Nur 100.000 wegen Laufzeit

Console.WriteLine("Sortiere 1 Million Elemente (au√üer Bubble Sort):");

// QuickSort
var sw = Stopwatch.StartNew();
QuickSort(dataForQuickSort, 0, dataForQuickSort.Length - 1);
sw.Stop();
Console.WriteLine($"QuickSort Dauer: {sw.ElapsedMilliseconds} ms");

// RadixSort
sw.Restart();
RadixSort(dataForRadixSort);
sw.Stop();
Console.WriteLine($"RadixSort Dauer: {sw.ElapsedMilliseconds} ms");

// BubbleSort (nur 100.000)
Console.WriteLine("\nBubbleSort mit 100.000 Elementen (zum Vergleich):");
sw.Restart();
BubbleSort(dataForBubbleSort);
sw.Stop();
Console.WriteLine($"BubbleSort Dauer: {sw.ElapsedMilliseconds} ms");

// üî¢ Zufallsdaten erzeugen
static int[] GenerateRandomArray(int size, int seed)
{
    var rand = new Random(seed);
    int[] array = new int[size];
    for (int i = 0; i < size; i++)
        array[i] = rand.Next(0, 1_000_000);
    return array;
}

// ‚ö° QuickSort
static void QuickSort(int[] array, int left, int right)
{
    if (left >= right) return;
    int pivot = array[(left + right) / 2];
    int index = Partition(array, left, right, pivot);
    QuickSort(array, left, index - 1);
    QuickSort(array, index, right);
}

static int Partition(int[] array, int left, int right, int pivot)
{
    while (left <= right)
    {
        while (array[left] < pivot) left++;
        while (array[right] > pivot) right--;
        if (left <= right)
        {
            (array[left], array[right]) = (array[right], array[left]);
            left++; right--;
        }
    }
    return left;
}

// üîÅ Radix Sort
static void RadixSort(int[] array)
{
    int max = GetMax(array);
    for (int exp = 1; max / exp > 0; exp *= 10)
        CountingSortByDigit(array, exp);
}

static void CountingSortByDigit(int[] array, int exp)
{
    int n = array.Length;
    int[] output = new int[n];
    int[] count = new int[10];

    for (int i = 0; i < n; i++)
        count[(array[i] / exp) % 10]++;

    for (int i = 1; i < 10; i++)
        count[i] += count[i - 1];

    for (int i = n - 1; i >= 0; i--)
    {
        int digit = (array[i] / exp) % 10;
        output[count[digit] - 1] = array[i];
        count[digit]--;
    }

    for (int i = 0; i < n; i++)
        array[i] = output[i];
}

static int GetMax(int[] array)
{
    int max = array[0];
    foreach (int num in array)
        if (num > max) max = num;
    return max;
}

// ü´ß Bubble Sort
static void BubbleSort(int[] array)
{
    int n = array.Length;
    bool swapped;

    for (int i = 0; i < n - 1; i++)
    {
        swapped = false;
        for (int j = 0; j < n - i - 1; j++)
        {
            if (array[j] > array[j + 1])
            {
                (array[j], array[j + 1]) = (array[j + 1], array[j]);
                swapped = true;
            }
        }
        if (!swapped)
            break;
    }
}
````
