## Dynamische Datenstrukturen

Dynamische Datenstrukturen sind Speicherstrukturen, die zur Laufzeit wachsen oder schrumpfen können. Im Gegensatz zu Arrays haben sie keine feste Größe. Sie erlauben eine flexible Speicherverwaltung und sind Grundlage vieler Algorithmen und Softwarearchitekturen.

Die meisten dynamischen Datenstrukturen kennen wir schon aus der Anwendung der Datenstrukturen in C#, die wir im Abschnitt [Einführung in komplexe Datenstrukturen](../CSharp_Introduction/Complex_Datastructures.de.ipynb) behandelt haben. Hier fassen wir die wichtigsten Konzepte zusammen und erweitern sie um einige Details.

### Listen (Linked Lists)

#### Arten:

* **Einfache verkettete Liste**: Jeder Knoten zeigt auf den nächsten.
* **Doppelt verkettete Liste**: Jeder Knoten zeigt auf den vorherigen und nächsten.
* **Zirkuläre Liste**: Die letzte zeigt wieder auf die erste.

#### Operationen:

* Hinzufügen am Anfang/Ende
* Entfernen von Elementen
* Durchlaufen (Iteration)

#### Vorteile:

* Dynamische Größe
* Einfache Einfüge-/Löschoperationen

#### Nachteile:

* Kein direkter Indexzugriff
* Overhead durch Zeiger

### Stack (LIFO – Last In First Out)

#### Prinzip:

* Das zuletzt eingefügte Element wird als erstes entfernt.

#### Operationen:

* `Push()`: Hinzufügen
* `Pop()`: Entfernen
* `Peek()`: Das oberste Element anzeigen

#### Anwendungen:

* Rückgängig-Funktion (Undo)
* Funktionsaufruf-Stack
* Klammerausdrücke parsen

### Queue (FIFO – First In First Out)

#### Prinzip:

* Das zuerst eingefügte Element wird als erstes entfernt.

#### Varianten:

* **Normale Queue**
* **Deque**: Doppelt-endige Queue (von beiden Seiten möglich)
* **Priority Queue**

#### Operationen:

* `Enqueue()`: Einfügen
* `Dequeue()`: Entfernen
* `Peek()`: Erstes Element anzeigen

#### Anwendungen:

* Druckerwarteschlangen
* Jobplanung
* Simulationen

### Bäume

#### Arten:

* **Binärbaum**: Jeder Knoten hat max. 2 Kinder
* **Binärsuchbaum (BST)**: Linke Knoten < Wurzel < rechte Knoten
* **AVL- / Rot-Schwarz-Bäume**: Selbstbalancierend
* **Heap**: Spezialbaum für Priority Queues
* **Trie**: Für Wörterbuch und Präfix-Suchen

#### Anwendungen:

* Hierarchien (Dateisysteme)
* Suchalgorithmen
* Parser und Expression Trees

### Hashtabellen (Hash Maps)

#### Prinzip:

* Schlüssel-Wert-Paare werden über eine Hashfunktion in einem Array abgelegt.

#### Operationen:

* `Insert(Key, Value)`
* `Get(Key)`
* `Delete(Key)`

#### Kollisionsstrategien:

* Chaining (verkettete Listen)
* Open Addressing (Linear / Quadratic Probing)

#### Vorteile:

* Sehr schnelle Suche (Ø O(1))
* Ideal für Caching, Lookups

#### Nachteile:

* Nicht sortiert
* Schlechte Hashfunktionen führen zu Kollisionen

### Dictionaries (Wörterbücher)

Ein **Dictionary** ist eine abstrakte Datenstruktur, die auf Hashtabellen basiert. In den meisten Programmiersprachen (z. B. `Dictionary<K, V>` in C#, `dict` in Python, `Map` in Java) ist es die konkrete Implementierung eines Key-Value-Stores.

#### Eigenschaften:

* **Schlüssel (Key)**: eindeutig, oft string oder int
* **Wert (Value)**: beliebiger Datentyp
* Intern häufig über **Hashtabellen** realisiert

#### Beispiel (C#):

```csharp
var telefonbuch = new Dictionary<string, string>();
telefonbuch["Anna"] = "01234";
telefonbuch["Ben"] = "05678";
Console.WriteLine(telefonbuch["Anna"]); // Ausgabe: 01234
```

#### Operationen:

* `Add(K, V)`: Fügt ein Paar hinzu
* `TryGetValue(K, out V)`: Sicheres Nachschlagen
* `ContainsKey(K)`: Existenzprüfung
* `Remove(K)`: Löschen eines Eintrags

#### Vorteile:

* Sehr schneller Zugriff auf Werte über Schlüssel (Ø O(1))
* Flexibler Einsatz in vielen Anwendungsfällen
* Lesbar und intuitiv (besonders in Sprachen wie Python)

#### Anwendungsfälle:

* Konfigurationen (z. B. "timeout": 3000)
* Zähler / Frequenz-Tabellen
* Objektzuordnung (z. B. `PersonId → PersonObjekt`)

#### Unterschiede zu einfachen Hashtabellen:

| Feature                | Hash Table (Low-Level) | Dictionary (High-Level)        |
| ---------------------- | ---------------------- | ------------------------------ |
| Benutzerfreundlichkeit | Niedrig                | Hoch                           |
| Typsicherheit          | Manuell                | Generics / dynamisch           |
| Zusätzliche Methoden   | Wenige                 | Viele                          |
| Fehlerbehandlung       | Selbst gemacht         | Integriert (z. B. TryGetValue) |

Damit ergänzen Dictionaries das Konzept von Hashtabellen um Typsicherheit, Methodenvielfalt und Komfortfunktionen – und sind in der Praxis das Mittel der Wahl für Key-Value-Datenhaltung.

### Landau-Symbole (Big-O-Notation)

Diese geben an, wie der Aufwand einer Operation in Bezug auf die Eingabemenge **n** wächst:

| Symbol     | Bedeutung     | Beispieloperation         |
| ---------- | ------------- | ------------------------- |
| O(1)       | konstant      | Zugriff auf Array-Index   |
| O(log n)   | logarithmisch | Binäre Suche              |
| O(n)       | linear        | Einfaches Durchlaufen     |
| O(n log n) | „log-linear“  | Mergesort                 |
| O(n²)      | quadratisch   | Doppelte Schleife         |
| O(2ⁿ)      | exponentiell  | Rekursive Fibonacci       |
| O(n!)      | faktoriell    | Brute-Force Permutationen |

#### 💡 Warum wichtig?

Sie helfen dir, die Skalierbarkeit deiner Algorithmen zu beurteilen – insbesondere bei großen Datenmengen.

### Zusammenfassung: Welche Struktur wann?

| Struktur    | Gut für…                              | Aufwand (Such-/Insert)     |
| ----------- | ------------------------------------- | -------------------------- |
| Liste       | Dynamische Sammlung, einfache Nutzung | O(n) / O(1) (am Anfang)    |
| Stack       | Rückverfolgung, Undo, Rekursion       | O(1)                       |
| Queue       | Warteschlangen, Scheduling            | O(1)                       |
| Baum (BST)  | Sortierte Daten, Suchen               | Ø O(log n), worst O(n)     |
| Hashtabelle | Schneller Zugriff auf Schlüssel       | Ø O(1), worst O(n)         |
| Trie        | Textsuche, Autovervollständigung      | O(k), k = Länge des Wortes |

## Sortieralgorithmen

Sortieralgorithmen sind ein zentraler Bestandteil der Informatik, insbesondere im Kontext dynamischer Datenstrukturen und der Big-O-Notation. Sie helfen dabei, Daten in eine bestimmte Reihenfolge zu bringen – meist aufsteigend oder absteigend. Das beeinflusst **Suchzeit, Datenvisualisierung, und weitere Verarbeitungsschritte** massiv.

### Überblick: Warum Sortieren?

* Vorbereitung für binäre Suche (z. B. O(log n) statt O(n))
* Effizientere Datenverarbeitung (z. B. Duplikaterkennung)
* Bessere Speicher- oder Netzwerkzugriffe (lokalitätserhaltend)
* Grundlage für viele Algorithmen (z. B. Medianberechnung)

### Klassifikation von Sortieralgorithmen

| Kriterium             | Typen / Beispiele                                             |
| --------------------- | ------------------------------------------------------------- |
| **Komplexität**       | O(n²): Bubble, Insertion<br>O(n log n): Merge, Heap, Quick    |
| **Stabilität**        | Stabil: MergeSort, InsertionSort<br>Instabil: QuickSort, Heap |
| **In-Place**          | Ja: QuickSort, HeapSort<br>Nein: MergeSort                    |
| **Vergleichsbasiert** | Ja: Alle klassischen Sorts<br>Nein: CountingSort, RadixSort   |
| **Adaptiv**           | Ja: Timsort, InsertionSort (gut für fast sortierte Listen)    |

### Wichtige Sortieralgorithmen im Vergleich

| Algorithmus      | Beste Zeit | Durchschnitt | Schlechteste Zeit | Speicher | Stabil |
| ---------------- | ---------- | ------------ | ----------------- | -------- | ------ |
| Bubble Sort      | O(n)       | O(n²)        | O(n²)             | O(1)     | ✅      |
| Insertion Sort   | O(n)       | O(n²)        | O(n²)             | O(1)     | ✅      |
| Selection Sort   | O(n²)      | O(n²)        | O(n²)             | O(1)     | ❌      |
| Merge Sort       | O(n log n) | O(n log n)   | O(n log n)        | O(n)     | ✅      |
| Quick Sort       | O(n log n) | O(n log n)   | O(n²)             | O(log n) | ❌      |
| Heap Sort        | O(n log n) | O(n log n)   | O(n log n)        | O(1)     | ❌      |
| Counting Sort    | O(n + k)   | O(n + k)     | O(n + k)          | O(k)     | ✅      |
| Radix Sort       | O(n·k)     | O(n·k)       | O(n·k)            | O(n + k) | ✅      |
| TimSort (Python) | O(n)       | O(n log n)   | O(n log n)        | O(n)     | ✅      |

### Sortieren und dynamische Strukturen

| Struktur        | Typisches Sortierverfahren    | Bemerkung                                    |
| --------------- | ----------------------------- | -------------------------------------------- |
| **Array**       | QuickSort, MergeSort, Timsort | Direkter Indexzugriff                        |
| **LinkedList**  | MergeSort                     | Kein direkter Zugriff; QuickSort ineffizient |
| **Binary Tree** | In-Order Traversierung        | Gibt die Elemente sortiert aus               |
| **HashTable**   | Nicht direkt sortierbar       | Sortieren erfordert Umkopieren               |
| **Heap**        | HeapSort                      | Nutzt die Heap-Eigenschaft (max/min)         |

### QuickSort

QuickSort ist ein sehr effizienter, rekursiver Sortieralgorithmus, der das "Teile und herrsche"-Prinzip nutzt. Er wählt ein **Pivot-Element** und partitioniert das Array in zwei Teile: Elemente kleiner als das Pivot und Elemente größer oder gleich dem Pivot. Diese Teile werden dann rekursiv sortiert.

<iframe width="768" height="400" src="https://visualgo.net/en/sorting?create=14%2C23%2C10%2C10%2C8%2C22%2C35%2C41%2C46%2C12&mode=Quick" frameborder="0" scrolling="no" allow="fullscreen; clipboard-read; clipboard-write" allowfullscreen></iframe>
</details>

#### Beispiel: QuickSort (rekursiv in C#)

```csharp
public static void QuickSort(int[] array, int left, int right)
{
    if (left >= right) return;

    int pivot = array[(left + right) / 2];
    int index = Partition(array, left, right, pivot);
    QuickSort(array, left, index - 1);
    QuickSort(array, index, right);
}

private static int Partition(int[] array, int left, int right, int pivot)
{
    while (left <= right)
    {
        while (array[left] < pivot) left++;
        while (array[right] > pivot) right--;
        if (left <= right)
        {
            (array[left], array[right]) = (array[right], array[left]);
            left++;
            right--;
        }
    }
    return left;
}
```

### Merksätze

* **O(n log n)** ist das Beste, was ein **vergleichsbasierter** Sortieralgorithmus im Durchschnitt leisten kann.
* **Stabilität** ist wichtig bei zusammengesetzten Schlüsseln.
* **In-place** ist wichtig bei großen Datenmengen mit begrenztem Speicher.
* Nicht-vergleichsbasierte Sortierverfahren sind für Integer- oder fix formatierte Daten extrem schnell.

### Radix Sort

Hier ist eine vollständige, kommentierte **Radix Sort**-Implementierung in **C#**, die auf Ganzzahlen basiert. Radix Sort funktioniert am besten bei positiven Ganzzahlen und nutzt die **Ziffernpositionen** (Einer, Zehner, Hunderter usw.), um mehrfach zu sortieren – jeweils mit einem stabilen Verfahren wie Counting Sort.

<iframe width="768" height="768" src="https://visualgo.net/en/sorting?create=3829%2C9231%2C5853%2C9755%2C3096%2C195%2C4044%2C991%2C944%2C48&mode=Radix" frameborder="0" scrolling="no" allow="fullscreen; clipboard-read; clipboard-write" allowfullscreen></iframe>
</details>

#### Beispiel: Radix Sort in C\#

```csharp
using System;
using System.Collections.Generic;

class Program
{
    static void Main()
    {
        int[] array = { 170, 45, 75, 90, 802, 24, 2, 66 };

        Console.WriteLine("Unsortiert:");
        PrintArray(array);

        RadixSort(array);

        Console.WriteLine("Sortiert:");
        PrintArray(array);
    }

    // Hauptmethode für Radix Sort
    public static void RadixSort(int[] array)
    {
        int max = GetMax(array);

        // Sortiere für jede Stellenposition: 1 (Einer), 10 (Zehner), 100 (Hunderter), ...
        for (int exp = 1; max / exp > 0; exp *= 10)
        {
            CountingSortByDigit(array, exp);
        }
    }

    // Ziffernbasierter Counting Sort (stabil)
    private static void CountingSortByDigit(int[] array, int exp)
    {
        int n = array.Length;
        int[] output = new int[n];
        int[] count = new int[10]; // Ziffern 0-9

        // Zähle Häufigkeit jeder Ziffer an aktueller Stelle
        for (int i = 0; i < n; i++)
        {
            int digit = (array[i] / exp) % 10;
            count[digit]++;
        }

        // Errechne kumulierte Positionen
        for (int i = 1; i < 10; i++)
        {
            count[i] += count[i - 1];
        }

        // Baue das Output-Array (von hinten für Stabilität)
        for (int i = n - 1; i >= 0; i--)
        {
            int digit = (array[i] / exp) % 10;
            output[count[digit] - 1] = array[i];
            count[digit]--;
        }

        // Kopiere zurück ins Originalarray
        for (int i = 0; i < n; i++)
        {
            array[i] = output[i];
        }
    }

    // Hilfsfunktion: finde das größte Element
    private static int GetMax(int[] array)
    {
        int max = array[0];
        foreach (int num in array)
        {
            if (num > max) max = num;
        }
        return max;
    }

    // Ausgabehilfe
    private static void PrintArray(int[] array)
    {
        foreach (int num in array)
        {
            Console.Write(num + " ");
        }
        Console.WriteLine();
    }
}
```

#### Was ist besonders an Radix Sort?

* **Nicht-vergleichsbasiert**: Nutzt die Struktur der Zahlen, nicht ihre Größe.
* **Stabil**: Wichtige Voraussetzung für korrektes Sortieren in mehreren Durchgängen.
* **Laufzeit**:

  * O(d·(n + k)), wobei `d` die Anzahl der Stellen, `n` die Anzahl der Elemente, und `k` der Zahlenbereich der Ziffern ist (hier 0–9 → konstant).
  * Für feste `d` ≈ O(n).

### Bubble Sort

Bubble Sort ist ein einfacher, aber ineffizienter Sortieralgorithmus, der oft als Lehrbeispiel verwendet wird. Er funktioniert, indem benachbarte Elemente wiederholt verglichen und vertauscht werden, bis die Liste sortiert ist. Der Name kommt von der Art und Weise, wie größere Elemente "nach oben steigen" (wie Blasen in Wasser).

</details>

Hier ist ein einfaches und gut kommentiertes Beispiel für **Bubble Sort** in **C#** – ideal, um das Funktionsprinzip zu verstehen und zu demonstrieren.

#### Bubble Sort in C\#

```csharp
using System;

class Program
{
    static void Main()
    {
        int[] array = { 5, 2, 9, 1, 5, 6 };

        Console.WriteLine("Unsortiert:");
        PrintArray(array);

        BubbleSort(array);

        Console.WriteLine("Sortiert:");
        PrintArray(array);
    }

    // 🫧 Bubble Sort Methode
    public static void BubbleSort(int[] array)
    {
        int n = array.Length;
        bool swapped;

        // Äußere Schleife für jeden Durchlauf
        for (int i = 0; i < n - 1; i++)
        {
            swapped = false;

            // Innere Schleife für paarweisen Vergleich
            for (int j = 0; j < n - i - 1; j++)
            {
                // Vergleiche benachbarte Elemente
                if (array[j] > array[j + 1])
                {
                    // Tausche, wenn falsch sortiert
                    (array[j], array[j + 1]) = (array[j + 1], array[j]);
                    swapped = true;
                }
            }

            // Wenn kein Tausch mehr erfolgt → sortiert
            if (!swapped)
                break;
        }
    }

    // 🖨 Hilfsfunktion zum Anzeigen des Arrays
    public static void PrintArray(int[] array)
    {
        foreach (int value in array)
        {
            Console.Write(value + " ");
        }
        Console.WriteLine();
    }
}
```

#### Eigenschaften von Bubble Sort

| Eigenschaft         | Wert                                |
| ------------------- | ----------------------------------- |
| Komplexität (Best)  | O(n) (wenn optimiert mit `swapped`) |
| Komplexität (Worst) | O(n²)                               |
| Speicherbedarf      | O(1) (in-place)                     |
| Stabil              | ✅ Ja                                |
| Einfach umzusetzen  | ✅ Ja                                |

#### Merksatz

> **Bubble Sort ist leicht zu verstehen, aber ineffizient für große Datenmengen.**
> Wird in der Praxis selten verwendet, außer zu Lehrzwecken.

## Anwendung: Vergleich von Sortieralgorithmen

Hier ist ein vollständiges **Beispiel in C#**, das **1 Million Zufallszahlen** generiert und sie mit den drei Sortieralgorithmen **Bubble Sort**, **Radix Sort** und **Quick Sort** sortiert – inklusive Zeitmessung.

> ⚠️ Hinweis: **Bubble Sort** ist extrem ineffizient bei dieser Menge. Deshalb wird er hier **auf nur 100.000 Elementen** angewendet, um das Programm nicht unnötig lange laufen zu lassen.

### Programm: Vergleich von Sortieralgorithmen mit 1 Mio. Testdaten

````polyglot-notebook
using System;
using System.Diagnostics;

const int N = 1_000_000;

// Testdaten generieren
int[] dataForQuickSort = GenerateRandomArray(N, seed: 42);
int[] dataForRadixSort = (int[])dataForQuickSort.Clone();
int[] dataForBubbleSort = GenerateRandomArray(100_000, seed: 42); // Nur 100.000 wegen Laufzeit

Console.WriteLine("Sortiere 1 Million Elemente (außer Bubble Sort):");

// QuickSort
var sw = Stopwatch.StartNew();
QuickSort(dataForQuickSort, 0, dataForQuickSort.Length - 1);
sw.Stop();
Console.WriteLine($"QuickSort Dauer: {sw.ElapsedMilliseconds} ms");

// RadixSort
sw.Restart();
RadixSort(dataForRadixSort);
sw.Stop();
Console.WriteLine($"RadixSort Dauer: {sw.ElapsedMilliseconds} ms");

// BubbleSort (nur 100.000)
Console.WriteLine("\nBubbleSort mit 100.000 Elementen (zum Vergleich):");
sw.Restart();
BubbleSort(dataForBubbleSort);
sw.Stop();
Console.WriteLine($"BubbleSort Dauer: {sw.ElapsedMilliseconds} ms");

// 🔢 Zufallsdaten erzeugen
static int[] GenerateRandomArray(int size, int seed)
{
    var rand = new Random(seed);
    int[] array = new int[size];
    for (int i = 0; i < size; i++)
        array[i] = rand.Next(0, 1_000_000);
    return array;
}

// ⚡ QuickSort
static void QuickSort(int[] array, int left, int right)
{
    if (left >= right) return;
    int pivot = array[(left + right) / 2];
    int index = Partition(array, left, right, pivot);
    QuickSort(array, left, index - 1);
    QuickSort(array, index, right);
}

static int Partition(int[] array, int left, int right, int pivot)
{
    while (left <= right)
    {
        while (array[left] < pivot) left++;
        while (array[right] > pivot) right--;
        if (left <= right)
        {
            (array[left], array[right]) = (array[right], array[left]);
            left++; right--;
        }
    }
    return left;
}

// 🔁 Radix Sort
static void RadixSort(int[] array)
{
    int max = GetMax(array);
    for (int exp = 1; max / exp > 0; exp *= 10)
        CountingSortByDigit(array, exp);
}

static void CountingSortByDigit(int[] array, int exp)
{
    int n = array.Length;
    int[] output = new int[n];
    int[] count = new int[10];

    for (int i = 0; i < n; i++)
        count[(array[i] / exp) % 10]++;

    for (int i = 1; i < 10; i++)
        count[i] += count[i - 1];

    for (int i = n - 1; i >= 0; i--)
    {
        int digit = (array[i] / exp) % 10;
        output[count[digit] - 1] = array[i];
        count[digit]--;
    }

    for (int i = 0; i < n; i++)
        array[i] = output[i];
}

static int GetMax(int[] array)
{
    int max = array[0];
    foreach (int num in array)
        if (num > max) max = num;
    return max;
}

// 🫧 Bubble Sort
static void BubbleSort(int[] array)
{
    int n = array.Length;
    bool swapped;

    for (int i = 0; i < n - 1; i++)
    {
        swapped = false;
        for (int j = 0; j < n - i - 1; j++)
        {
            if (array[j] > array[j + 1])
            {
                (array[j], array[j + 1]) = (array[j + 1], array[j]);
                swapped = true;
            }
        }
        if (!swapped)
            break;
    }
}
````
