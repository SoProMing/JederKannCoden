### Top-down vs. Bottom-up Design

Diese beiden Begriffe bezeichnen **entgegengesetzte Denk- und Vorgehensweisen** beim Entwurf von Systemen – sei es bei der Softwarearchitektur, beim Programmieren, bei der Datenmodellierung oder in der allgemeinen Systementwicklung.

### Top-down Design (auch: abstrakt nach konkret)

#### Grundidee

Beim Top-down-Ansatz beginnt man **auf der höchsten Abstraktionsebene**: mit einer Vision, einem Ziel oder einem System als Ganzes. Dieses wird in **Teilkomponenten zerlegt**, bis man bei konkreten, implementierbaren Modulen ankommt.

#### Vorgehensweise

1. Definition der Gesamtfunktionalität oder Aufgabe des Systems
2. Zerlegung in Hauptmodule (Makrostruktur)
3. Weitere Unterteilung in Subsysteme oder Funktionen
4. Detaillierung der Implementierung

#### Praxisbeispiel

Beim Entwurf eines **Navigationssystems**:

* Ziel: „Ein Gerät, das Nutzer von A nach B navigiert.“
* Unterteilung: Benutzerinterface, Kartenmodul, Routenberechnung, GPS-Empfänger
* Weiter: Routenberechnung → Dijkstra-Algorithmus → Datenstrukturen etc.

#### Hilfsmittel

* **Struktogramme**
* **UML Use-Case Diagramme**, Aktivitätsdiagramme
* **Flowcharts**
* **Pseudocode**
* **Decomposition Trees**

#### ✅ Vorteile

* Gute Übersicht von Anfang an
* Frühzeitige Spezifikation der Anforderungen
* Klare Aufteilung der Verantwortlichkeiten

#### ⚠️ Nachteile

* Gefahr der „Overengineering“
* Fehlende Wiederverwendung von bereits existierenden Komponenten
* Änderungen auf hoher Ebene können weitreichend sein

### Bottom-up Design (auch: von konkret zu abstrakt)

#### Grundidee

Hier beginnt man **mit konkreten Bausteinen oder Modulen**, die gut verstanden oder bereits vorhanden sind. Diese werden dann **zu größeren Systemen zusammengefügt**.

#### Vorgehensweise

1. Identifikation oder Erstellung wiederverwendbarer Basismodule (z. B. Klassen, Funktionen)
2. Kombination dieser zu komplexeren Strukturen
3. Aufbau höherer Abstraktionsebenen
4. Integration zum Gesamtsystem

#### Praxisbeispiel

Entwicklung eines **Finanzsystems**:

* Man beginnt mit der Implementierung von Grundfunktionen: „Zinsen berechnen“, „Zahlung durchführen“, „Kontostand abfragen“.
* Diese Funktionen werden kombiniert zu Modulen: „Konto“, „Buchhaltung“, „Reporting“.
* Daraus entsteht das Gesamtprodukt.

#### Hilfsmittel

* **Unit Tests**
* **Modulbibliotheken**
* **Abstrakte Datentypen (ADT)**
* **Klassen und Objektmodelle**
* **Refactoring-Werkzeuge**

#### ✅ Vorteile

* Wiederverwendung bestehender Lösungen
* Flexibler bei Änderungen
* Frühzeitige Implementierung und Testbarkeit

#### ⚠️ Nachteile

* Gefahr von Inkonsistenzen in der Gesamtstruktur
* Keine klare Gesamtarchitektur zu Beginn
* Schwerer zu koordinieren bei größeren Teams

### Vergleich

| Aspekt           | Top-down                      | Bottom-up                         |
| ---------------- | ----------------------------- | --------------------------------- |
| Ausgangspunkt    | Gesamtproblem                 | Einzelkomponenten                 |
| Denkweise        | Abstrakt → Konkret            | Konkret → Abstrakt                |
| Planung          | Starke Vorplanung             | Iteratives Wachsen                |
| Flexibilität     | Eher gering                   | Hoch                              |
| Wiederverwendung | Eher gering (außer geplant)   | Hoch (Nutzung vorhandener Module) |
| Eignung          | Große, klar umrissene Systeme | Explorative, modulare Systeme     |

### In der Praxis: Kombination beider Ansätze

<a href="https://miro.com/app/board/o9J_lOJi2o0=/?moveToWidget=3074457357770715523&cot=14"><img src="Top down bottom up software design.jpg" style="width:100%;"></a>

In modernen Softwareprojekten wird oft **beides kombiniert**:

* Man beginnt **Top-down**, um das große Ganze zu verstehen.
* Parallel oder danach werden Komponenten **Bottom-up** entwickelt.
* Dieser Ansatz ist z. B. bei **Agile**, **Domain-Driven Design** und **Clean Architecture** üblich.

### Hilfsmittel und Methoden in der Praxis

| Tool / Methode                | Unterstützt …         | Beschreibung                                        |
| ----------------------------- | --------------------- | --------------------------------------------------- |
| UML (Klassendiagramme etc.)   | beide                 | Visualisierung von Struktur und Beziehungen         |
| Clean Architecture            | beide (innen ↔ außen) | Layered Design, lose Kopplung                       |
| Test Driven Development (TDD) | Bottom-up             | Fokus auf kleine, getestete Module                  |
| Domain Driven Design (DDD)    | Top-down              | Strategisches Design nach Fachdomäne                |
| Prototyping                   | Bottom-up             | Frühzeitiges Ausprobieren und Kombinieren           |
| Scrum / Agile Boards          | beide                 | Unterstützen schrittweisen Aufbau und Visionserhalt |

### Merksatz

>
> **Top-down**: „Ich plane ein Haus und baue es nach Plan.“
>
> **Bottom-up**: „Ich habe Ziegel, Fenster, Balken – daraus baue ich ein Haus.“
